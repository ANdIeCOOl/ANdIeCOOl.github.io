```python
duration_series = []
for x in df["Duration"]:
  #print(x)
  #print(type(x))
  h = 0
  m = 0
  l = x.split(" ")
  for y in l:
    if "h" in y:
      h += int(''.join(map(str, (y[:-1]))))

    if "m" in y:
      # print(type(y[:-1]))
      # print(y[:-1])
      m += int(''.join(map(str, (y[:-1]))))
      #y += int
  seconds = h*60*60 + m*60
  duration_series.append(seconds)
```

```python
import ast

def extract_directors(crew):
  try:
    crew_list = ast.literal_eval(crew)
    directors = [person["name"] for person in crew_list if person["job"] == "Director"]
    return directors
  except(ValueError, SyntaxError):
    return []

df_merged["directors"] = df_merged["crew"].apply(extract_directors)
```

```python
C = df_merged["vote_average"].mean() #avg across whole report
m = df_merged["vote_count"].quantile(0.9) #min threshold

def weighted_rating(x,m=m, C=C):
  if (x["vote_count"] < m):
    return np.nan
  else:
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)
df_merged["weighted_rating"] = df_merged.apply(weighted_rating, axis=1)
```
